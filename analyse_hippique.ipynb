### Gestion variable qualitative

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeCV, LassoCV, Ridge, Lasso
from sklearn.linear_model import ElasticNet, SGDClassifier
```

```python
def dummies(dataset, column):
    dummies = pd.get_dummies(dataset,prefix=column, drop_first= True)
    dataset = pd.concat([dataset,dummies],axis=1)
    return dataset
```

```python
df_reunion = dummies(df, 'cr-reunion')
df_num = dummies(df, 'cr-num')
df_Evt = dummies(df, 'cr-Evt')
df_autostart = dummies(df, 'cr-autostart')
df_corde = dummies(df, 'cr-corde')
df_etat_du_terrain = dummies(df, 'cr-etat du terrain')
df_distance = dummies(df, 'cr-distance')
df_sexe = dummies(df, 'Sexe')
```

```python
df_date = dummies(df, "cr-Date")
#df["cr-Date"].value_counts()
```

```python
df_hippo = pd.get_dummies(df['cr-Hippodrome'], prefix = 'cr-Hippodrome')
df = pd.concat([df,df_hippo],axis=1)
```

```python
df = df.drop(['cr-Hippodrome'], axis = 1)
```

### drop des colonnes get dummies

```python
X = df.drop(["place", "cr-Date","cr-reunion", "cr-num", "cr-Evt", "cr-autostart", "cr-corde", "cr-etat du terrain", "cr-distance", "Sexe" ],1)   #Feature Matrix
y = df["place"]          #Target Variable
```

### Training lasso sur notre dataset

```python
reg = LassoCV()
reg.fit(X, y)
print("Best alpha using built-in LassoCV: %f" % reg.alpha_)
print("Best score using built-in LassoCV: %f" %reg.score(X,y))
coef = pd.Series(reg.coef_, index = X.columns)
```

```python
print("Lasso a retenu " + str(sum(coef != 0)) + " variables et Ã©liminÃ© " +  str(sum(coef == 0)) + " variables")
```

```python
imp_coef = coef.sort_values()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (50.0, 50.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Lasso Model")
```
